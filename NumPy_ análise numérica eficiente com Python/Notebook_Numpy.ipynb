O coenficiente angular (a) pode ser obtido através da fórmula:

image.png

```python
Y = Moscow
X = datas
n = np.size(Moscow) # numero de elementos dentro de um array
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y)) / (n*np.sum(X**2) - np.sum(X)**2)
a
```

O coeficiente linear (b) pode ser obtido através da fórmula:

image.png

```python
b = np.mean(Y) - a * np.mean(X)
b
```

```python
y = a*X + b
```

```python
np.linalg.norm(Moscow - y)
```

```python
plt.plot(datas, Moscow)
plt.plot(x, y)
plt.plot(41.5, 41.5*a+b, '*r') # ponto que queremos observar, *r gera uma estrelinha e define a cor dela para vermelho
plt.plot(100, 100*a+b, '*r') # como calculamos os coeficientes que prevem o crescimento da reta, podemos usar isso para fazer previsões de pontos que ainda não temos dados
```

```python
coef_angulares = np.random.uniform(low = 0.10, high = 0.90, size = 100) # mesma coisa da função anterior porém para números float
```

```python
norma = np.array([]) # cria um array vazio

for i in range(100):
  norma = np.append(norma, np.linalg.norm(Moscow - (coef_angulares[i]*X + b)))
```

```python
min = norma[0]
index = 0

for i in range(100):
  if norma[i] < min:
    min = norma[i]
    index = i

coef_angulares[index] # gera um valor proximo ao que haviamos visto anteriormente
```

```python
coef_angulares[norma == np.min(norma)] # pode ser feito de maneira mais direta
```
