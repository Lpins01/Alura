```python
from sqlalchemy import create_engine, MetaData, Table, inspect
# create_engine -> cria conexão com o banco de dados
# MetaData -> representa os metadados do banco de dados, incluindo informações sobre tabelas, colunas e outros objetos
# Table -> usada para criar e manipular dados em uma tabela específica
# inspect -> função que permite inspecionar a estrutura do banco de dados, incluindo tabelas, colunas e outros objetos
```

```python
engine = create_engine('sqlite:///:memory')
# sqlite -> já está instalado no google colab, não precisando de um servidor separado
# memory -> representa que o banco de dados será local
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'

dados = pd.read_csv(url)
```

```python
dados.head()
```

```python
dados.to_sql('clientes', engine, index = False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"' # quando passamos o parâmetro que seria o valor (string) que a coluna deve ter, este deve estar em um tipo diferente de aspas da geral
```

```python
empregados = pd.read_sql(query, engine)
# read_sql é usado para fazer a leitura de uma consulta
```

```python
empregados
```

```python
empregados.to_sql('empregados', con = engine, index = False)
```

```python
pd.read_sql_table('empregados', engine)
# read_sql_table serve para fazer a leitura de uma tabela inteira
```

```python
pd.read_sql_table('empregados', engine, columns = ['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual']) # conseguimos especificar as colunas que queremos visualizar
```

```python
query = 'DELETE FROM clientes WHERE ID_Cliente = 5008804' # DELETE ajuda a remover uma informação
with engine.connect() as conn:
  conn.execute(query)

# with garante que o banco de dados será encerrado após a consulta
```
