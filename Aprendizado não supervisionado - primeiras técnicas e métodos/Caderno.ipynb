* O **Princípio APRIORI** nos diz que se um conjunto de itens é frequente, então todos os seus subconjuntos também são, fazendo com que os conjuntos de itens não frequentes e seus subconjuntos sejam podados

A imagem a seguir demostra um exemplo do algoritmo, de início são removidos (conjuntos em azul) os conjuntos que possuem suporte abaixo do mínimo estabelecido, a partir disso, os conjuntos possíveis que são gerados a partir dos dados que ainda são frequentes são avaliados e novamente os que estão abaixo do suporte mínimo são podados

image.png

```python
from mlxtend.frequent_patterns import apriori
```

A medida de Suporte estaria relacionada a quantidade de vezes que determinado item, ou conjunto de itens aparece no todo


*   Fração das transações em que ocorrem um conjunto de itens
*   Conjunto será frequente se estiver **acima do limiar** estipulado pelo(a) usuário(a)
*   **Importância:** verificar regras que ocorrem por coincidência, eliminar regras de baixo interesse para o negócio

A imagem a seguir mostra o suporte para uma regra, que mostraria a quantidade de vezes que o conjunto X e o conjunto Y aparecem juntos no total, nos ajudando a verificar se a compra do primeiro conjunto pode influenciar na compra do segundo.

image.png



```python
items_frequentes_apriori = apriori(transacao_transformado, use_colnames=True, min_support=0.02)
items_frequentes_apriori.sort_values(['support'], ascending=False)
```

```python
from mlxtend.frequent_patterns import association_rules
```

**Confiança**



*   Mede a frequência que os itens do conjunto Y aparecem em transações que contenham o conjunto X
*   **Importância:** mede a confiabilidade de uma inferência feita por uma regra

A imagem a seguir mostra o cálculo da confiança, esse cálculo nos ajuda a visualizar a quantidade de vezes que o conjunto X e o conjunto Y aparecem juntos dentre todas as vezes que o conjunto X aparece.

image.png

**Atenção!**

*   Cuidado ao interpretar as associações
*   A regra de associação indica uma **forte co-ocorrência** (se um item foi comprado existe a **possibilidade** do outro ser comprado) entre os intens antecedente (conjunto X) e consequente (conjunto Y), e não causalidade



```python
regras_apriori = association_rules(items_frequentes_apriori, metric='confidence', min_threshold=0.5)
regras_apriori
```

## Aula 2 - Clustering com K-means

### O que é Clustering (agrupamento)?

*   Identificação de grupos (clusters), a partir dos dados com características em comum, que sejam significativos e úteis

* "A análise de agrupamentos representa um conjunto de técnicas exploratórias utilizadas quando há a intenção de se verificar a existência de comportamentos semelhantes entre observações em relação a determinadas variáveis e o objetivo de se criarem grupos, ou clusters, em que prevaleça a homogeneidade interna."

* Dica de leitura: [Clustering - Biblioteca SKlearn](https://scikit-learn.org/stable/modules/clustering.html), [Capítulo 9 - Manual de Análise de Dados](https://www.google.com.br/books/edition/Manual_de_An%C3%A1lise_de_Dados/SmlaDwAAQBAJ?hl=pt-BR&gbpv=0), [Capítulo 9 - Mãos à obra: aprendizado de máquina com Scikit-Learn, Keras & TensorFlow: Conceitos, ferramentas e técnicas para a construção de sistemas inteligentes](https://www.google.com.br/books/edition/M%C3%A3os_A_Obra_Aprendizado_De_M%C3%A1quina_Com/JUBIEAAAQBAJ?hl=pt-BR&gbpv=0)


**Abordagens**

*   **Abordagem Hierárquica:** A ideia principal deste algoritmo é agrupar os dados utilizando uma hierarquia da relação entre eles, ou seja, criando uma árvore de clusters. Suponha que temos dois pontos nos dados e estes pertencem a grupos individuais. Por estarem próximos, se mesclam e se tornam um cluster, criando assim um tipo de hierarquia, como se um agrupamento estivesse dentro de outro agrupamento. Podemos citar dois algoritmos deste tipo: clustering aglomerativo (agglomerative clustering) e o BIRCH.

*   **Abordagem Baseada em Densidade:** Este tipo de algoritmo procura identificar as regiões com altas densidades (alta concentração) de pontos no espaço dos dados e agrupar como pertencentes ao mesmo cluster, podendo ter geometrias arbitrárias, ou seja, diferentes formas de cluster. Além disso, outliers podem ser identificados como aqueles pontos em que não foi possível realizar um agrupamento. Podemos citar o DBSCAN e o Mean Shift, como os algoritmos principais desta abordagem.

*   **Abordagem Baseada em Centróides:** Neste algoritmo, os dados são agrupados de acordo com a distância até o ponto central ou centróide do cluster. Sua aplicação possui muitas vantagens, como ser simples de implementar e computacionalmente eficiente. Porém, ele é sensível a outliers e às condições iniciais, como o valor de K-clusters para o algoritmo K-Means, que inclusive é o mais utilizado nesta abordagem.

### Aplicando o K-means



Algoritmo da **Análise de Agrupamento** que tenta encontrar um número especificado pelo(a) usuário(a) de clusters (K), que são representados por seus centróides.

**Vantagem:** Simples e capaz de identificar grupos com rapidez e eficiência

**Como funciona:**


*   Inicializar os K centróides
*   Calcular a distância de cada ponto com cada centróide e "rotular" (conectar) ao que estiver mais próximo
*   Atualizar o centróide, com a média de seus pontos
*   Repetir até que nenhum ponto mude mais ou nenhum centróide seja alterado

### Aplicando o Mean Shift

* **MeanShift**

    [Documentação Sklearn](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.MeanShift.html#sklearn.cluster.MeanShift)

    [Documentação `estimate_bandwidth`](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.estimate_bandwidth.html?highlight=bandwidth#sklearn.cluster.estimate_bandwidth)

* Dica de leitura: [Artigo científico - Mean shift: a robust approach toward feature space analysis](https://courses.csail.mit.edu/6.869/handouts/PAMIMeanshift.pdf)

* Método iterativo simples que desloca cada centróide para ser a média dos pontos de dados em uma determinada região.
* Os centróides serão deslocados até convergirem ao ponto de máxima densidade (grande concentração de pontos)
  * O ponto de máxima densidade é encontrado quando ao calcular a média das distâncias dos pontos do agrupamento até o centróide, esse centróide permanece na mesma posição da última interação, formando assim o primeiro cluster
* Podem haver níveis de agrupamentos
* Diferentes pontos de inicialização (diferentes pontos podem se conectar a diferentes grupos, formando clusters diferentes dependendo da situação)
* Atenção ao tamanho do raio da região! (determina quantos clusters vão ser identificados e qual o caminho de deslocamento do centróide)

* **Como funciona**
  * Inicialmente é definido um Bandwidth (região de interesse)
  * O centro dessa região será um centróide
  * O algoritmo então irá calcular a média das distâncias dos pontos dentro dessa região até o centróide
  * Essa média irá definir a nova localização do centróide
  * Os passos são repetidos até que a posição do centróide não seja mais modificada, encontrando assim a região de maior concentração de pontos (máxima densidade)

```python
from sklearn.cluster import MeanShift, estimate_bandwidth # vai estimar o tamanho da região de interesse ideal para os dados
```

```python
bw = estimate_bandwidth(X_normalizado)
```

```python
meanshift_dados = MeanShift(bandwidth=bw).fit(X_normalizado)
```

```python
rotulos_meanshift = meanshift_dados.labels_
numero_rotulos = len(np.unique(rotulos_meanshift)) # identifica a quantidade de rótulos únicos
numero_rotulos
```

```python
cluster_centers = meanshift_dados.cluster_centers_
cluster_centers
```

```python
sns.set(style='white', rc={'figure.figsize':(9,6)},font_scale=1.1)

#plot dos agrupamentos
plt.scatter(x=X_normalizado[:,0], y=X_normalizado[:, 1], c=rotulos_meanshift, cmap='winter')

#plot dos centroides
plt.scatter(meanshift_dados.cluster_centers_[:, 0],
            meanshift_dados.cluster_centers_[:, 1],
            s=100,
            c='red')

plt.title('Dados Agrupados com MeanShift',fontweight='bold')
plt.show()
```

### Aplicando DBSCAN ao problema

* **DBSCAN**

    [Documentação Sklearn](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html#sklearn.cluster.DBSCAN)

* Density-Based Spatial Clustering of Applications with Noise (Clusterização Espacial Baseada em Desnidade de APlicações com Ruído)
* Define clusters como regiões com alta densidade
  * Aquele ponto que não for agrupado com nenhum cluster será considerado um ruído, ou seja, será um dado que não pertencerá a nenhum grupo
* Não é necessário o número de clusters definido
* Atenção ao valor do epsilon (raio da região que irá se passar todos os pontos)
  * O epsilon define a distância máxima entre pontos para que um seja considerado pertencente a vizinhança do outro, ou seja, mesmo grupo
  * No **MeanShift**, os centróides eram novos pontos representados pelas médias das disntâncias, já no **DBSCAN**, os centróides serão os próprios pontos
* Pontos podem ser considerados ruídos (isso acontece muitas vezes por conta do valor de epsilon, ou seja, a região que estará sendo considerada para agrupar os dados)


* **Como funciona**
  * Inicialmente é selecionado um ponto qualquer para ser o centróide inicial e um epsilon (raio do grupo)
  * Esse ponto então é "marcado" como pertencente ao grupo e o ponto mais próximo deste que também esteja dentro daquele raio previamente definido será o novo centróide
  * Quando não existiram mais pontos para serem marcados dentro desse raio um novo centróide aleatório é definido até que todos os grupos sejam identificados.
  * Se um ponto não formar grupo com nenhum outro este é considerado um ruído

```python
from sklearn.cluster import DBSCAN
```

```python
#realizar varios testes com os parametros `eps` (raio da região) e `min_samples` (número mínimo de pontos para que seja considerado um agrupamento)
dbscan = DBSCAN(eps=1.20, min_samples=2)
```

```python
dbscan_dados = dbscan.fit(X_normalizado)
```

```python
rotulos_dbscan = dbscan_dados.labels_ # Os pontos que forem ruídos aparecerão como -1
rotulos_dbscan
```

```python
np.unique(rotulos_dbscan, return_counts=True) # Contagem de agrupamentos
# O prineuri array irá mostrar a quantidade de grupos (se existir algum com valor -1 indica ruído)
# O segundo array irá mostrar a quantidade de pontos em cada grupo
```

```python
numero_rotulos = len(np.unique(rotulos_dbscan))
numero_rotulos
```

```python
sns.set(style='white', rc={'figure.figsize':(9,6)},font_scale=1.1)

#plot dos agrupamentos
plt.scatter(x=X_normalizado[:,0], y=X_normalizado[:, 1], c=rotulos_dbscan, cmap='viridis')

plt.title('Dados Agrupados com DBSCAN',fontweight='bold')
plt.show()
```

## Aula 4 - Redução de dimensionalidade


### O que é Redução de Dimensionalidade?


* O que é Maldição da Dimensionalidade?

    Fenômeno que ocorre quando temos uma grande quantidade de variáveis e muitas vezes são altamente correlacionadas ou não agregam informação útil ao problema que estamos lidando.

* Redução de dimensionalidade

    Reduzir o número de características para que reduza o custo e a facilite o desenvolvimento de uma solução.
    
    Busca reduzir a quantidade de dimensões para um pequeno subconjunto que capture a maior parte da variação dos dados


### A Redução de Dimensionalidade por ser feita através de diversas técnicas

* **Análise de Componentes Principais (Principal Component Analysis - PCA):** Este algoritmo tem como objetivo criar novas features a partir da combinação linear das variáveis originais, explicando a variância máxima dos dados.

* **Kernel PCA:** Esta técnica é uma variação do PCA para lidar com dados não lineares. A vantagem consiste em utilizar o [método kernel](https://en.wikipedia.org/wiki/Kernel_method) para encontrar uma projeção dos dados que seja possível separá-los linearmente. A biblioteca Scikit-Learn possui ferramentas para implementar esta técnica. Confira neste [link](https://scikit-learn.org/stable/auto_examples/decomposition/plot_kernel_pca.html#sphx-glr-auto-examples-decomposition-plot-kernel-pca-py).

* **Análise Discriminante Linear (Linear Discriminant Analysis - LDA):** Na maioria dos casos é utilizado como um classificador linear (aprendizado supervisionado). No entanto, pode ser usado para reduzir a dimensão dos dados ao buscar uma projeção que maximize a separação de classes. Ou seja, a partir das variáveis originais, podemos extrair novas variáveis que separem ao máximo estes dados em classes. Assim como as demais técnicas, há na biblioteca Scikit-Learn ferramentas para implementar, pode ser conferido neste [link](https://scikit-learn.org/stable/modules/generated/sklearn.discriminant_analysis.LinearDiscriminantAnalysis.html).

* **Isomap:** Também conhecido como **Isometric mapping**, é um algoritmo que busca uma dimensão menor que mantenha as distâncias [geodésicas](https://pt.wikipedia.org/wiki/Geod%C3%A9sica) entre todos os pontos. Essas distâncias consideram variações em forma de curva e não uma linha reta de distância entre os pontos. Para entender melhor as aplicações desta técnica, confira a [documentação do sklearn](https://scikit-learn.org/stable/modules/manifold.html#isomap).

### O que é PCA?

* Faremos uma simulação desta técnica a seguir, utilizando o problema da **Indústria Pé de Feijão**!

* O que é PCA (Principal Component Analysis)?

  * Tem como objetivo selecionar uma ou mais variáveis que concentram a maior parte da variação do dados
  * Busca extrair uma combinação interpretável das variáveis originais (quando temos muitos dados podemos tentar eliminar variáveis redundantes e combinar as possíveis)
  * **Vantagens:**
    * Identificação de padrões
    * Eliminação de ruídos (ao combinar variáveis conseguimos eliminar informações insignificantes)
    * Variablidade explicada por uma dimensão reduzida

    "A análise fatorial por componentes principais permite, portanto, que, a partir de um conjunto de variáveis originais correlacionadas entre si, seja determinado outro conjunto de variáveis (fatores) resultantes da combinação linear do primeiro conjunto." - Livro: Manual de Análise de Dados de autoria de Luiz Paulo Fávero e Patrícia Belfiore.

* Dica de leitura: [Documentação Sklearn - Técnicas de Redução de Dimensionalidade](https://scikit-learn.org/stable/modules/decomposition.html#decompositions), [Capítulo 10 - Manual de Análise de Dados](https://www.google.com.br/books/edition/Manual_de_An%C3%A1lise_de_Dados/SmlaDwAAQBAJ?hl=pt-BR&gbpv=0)
